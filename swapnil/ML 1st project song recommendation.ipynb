```python
import pandas as pd
```

```python
triplets_file = 'https://static.turi.com/datasets/millionsong/10000.txt'
songs_metadata_file = 'https://static.turi.com/datasets/millionsong/song_data.csv'
```

```python
data=pd.read_table(triplets_file,header=None)  # TAKING 1st set of data containing user count......
```

```python
data
```

```python
data2=pd.read_csv(songs_metadata_file) # Taking second set of data having information of the songs.....
```

```python
data2
```

```python
overall_data=pd.merge(data,data2,on="song_id")     # combining both data on basis of song id ......
```

```python
overall_data
```

```python
modified_data=overall_data.dropna(axis=0,subset=["title"]) #  Removing Nan values from the title column
```

```python
modified_data=modified_data.dropna(axis=0,subset=["release"]) # Removing Nan "    "  "    "   release column
```

```python
modified_data=modified_data.dropna(axis=0,subset=["artist_name"]) # Removing Nan " "  " "   artist name column
```

```python
modified_data
```

```python
total_count = (modified_data.groupby(by=['song_id'])['count'].count().reset_index().rename(columns={'count':'total_count'})[['song_id','total_count']])
```

```python
total_count  # storing total counts of the data....
print(total_count['total_count'].describe()) # to see threshold for data....
```

```python
data_with_rating=overall_data.merge(total_count,left_on='song_id',right_on='song_id',how="left") ## Adding the total count to past data
```

```python
total_count
```

```python
data_with_rating # This data contains the total count additional column...
```

```python
data_table=data_with_rating.pivot_table(index='title',columns='user_id',values='count').fillna(0) # To use cosine transform ,converted to pivot table...
```

```python
data_table
```

```python
from scipy.sparse import csr_matrix  # using cosine transforms

song_features_matrix= csr_matrix(data_table.values)

from sklearn.neighbors import NearestNeighbors # using neighrest neighbors

model=NearestNeighbors(metric='cosine',algorithm='brute')
model.fit(song_features_matrix)
```

```python
import numpy as np
```

```python
any_user=np.random.choice(data_table.shape[0]) # taking any random song
print(any_user)
distances,indices =model.kneighbors(data_table.iloc[any_user,:].values.reshape(1,-1),n_neighbors=6) # use 6 neighrest neigbours of paricular song
```

```python
 for i in range(0,len(distances.flatten())):
        if i==0:
            print('Recommendation for {0}:\n'.format(data_table.index[any_user]))
        else:
            print('{0}:{1},with distances of {2}:'.format(i,data_table.index[indices.flatten()[i]],distances.flatten()[i])) # use 6 neighrest neigbours including the same movie
```
