```python
import pandas as pd
import numpy as np

triplets_file = r'C:\Users\Sushma\Downloads\labels.txt'
song_metadata_file = r'C:\Users\Sushma\Downloads\Million_Song_Subset.xlsx'
```

```python
song_df_1 = pd.read_table(triplets_file, header=None)
song_df_1.columns = ['user_id', 'song_id', 'listen_count']
song_df_2 = pd.read_excel(song_metadata_file)
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on="song_id", how="left")
song_df.head()
```

```python
print(len(song_df))
users = song_df['user_id'].unique()
print(len(users))
songs = song_df['song_id'].unique()
print(len(songs))
```

```python
modified_data = song_df.dropna(axis=0,subset=["title"])
modified_data = song_df.dropna(axis=0,subset=["release"])
modified_data = song_df.dropna(axis=0,subset=["artist_name"])
song_df
```

```python
complete_count = (modified_data.groupby(by=['song_id'])['listen_count'].count().reset_index().rename(columns={'listen_count':'total_count'})[['song_id','total_count']])
complete_count
```

```python
print(complete_count['total_count'].describe())
```

```python
rated_song_df=song_df.merge(complete_count,left_on='song_id', right_on='song_id', how="left")
complete_count
```

```python
rated_song_df
```

```python
song_table=rated_song_df.pivot_table(index='title', columns='user_id', values='listen_count').fillna(0)
song_table
```

```python
from scipy.sparse import csr_matrix
song_features_matrix=csr_matrix(song_table.values)
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='cosine',algorithm='brute')
model.fit(song_features_matrix)
```

```python
any_user=np.random.choice(song_table.shape[0])
print(any_user)
distances,indices = model.kneighbors(song_table.iloc[any_user,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendation for{0}:\n'.format(song_table.index[any_user]))
    else:
        print('{0}:{1},with distance of {2}:'.format(i,song_table.index[indices.flatten()[i]],distances.flatten()[i]))
```
